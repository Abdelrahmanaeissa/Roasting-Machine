```python
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
from scipy.stats import stats
from scipy.stats import norm
from statsmodels.stats import weightstats 
```

```python
x1 = df1['T_data_1_1']
mean_x1 = np.mean(x1)
std_x1 = np.std(x1)
```

```python
x2 = df1['T_data_1_2']
mean_x2 = np.mean(x2)
std_x2 = np.std(x2)
```

```python
x3 = df1['T_data_1_3']
mean_x3 = np.mean(x3)
std_x3 = np.std(x3)
```

```python
room1 = pd.concat([pd.DataFrame(data = {'Readings': x1, 'Sensor':'Sensor 1'}),pd.DataFrame(data = {'Readings': x2, 'Sensor':'Sensor 2'}),
                   pd.DataFrame(data = {'Readings': x3, 'Sensor':'Sensor 3'})])
                                                                                                                                                                                                                                                                                                      
```

```python
sns.displot(data = room1, x= 'Readings',bins = 50, col = 'Sensor')
plt.xlabel('Readings')
plt.show()
plt.tight_layout() 
```

```python
y1 = norm.pdf(x1, mean_x1, std_x1)
limit_1 =  (1.96 * std_x1) + mean_x1
limit_2 = mean_x1 - (1.96 * std_x1)
```

```python
y2 = norm.pdf(x2, mean_x2, std_x2)
limit_21 =  (1.96 * std_x2) + mean_x2
limit_22 = mean_x2 - (1.96 * std_x2)
```

```python
y3 = norm.pdf(x3, mean_x3, std_x3)
limit_31 =  (1.96 * std_x1) + mean_x3
limit_32 = mean_x3 - (1.96 * std_x3)
```

```python
sns.lineplot(x = x1, y= y1)
plt.axvline(x=mean_x1, color = 'black')
plt.axvline(x=limit_1, color = 'red')
plt.axvline(x=limit_2, color = 'red')
plt.xticks([np.min(x1),limit_1,limit_2, np.max(x1)])
plt.ylabel('probability')
plt.xlabel('readings')
plt.title('Sensor 1 Distribution')
plt.show()
```

```python
sns.lineplot(x = x2, y= y2)
plt.axvline(x=mean_x2, color = 'black')
plt.axvline(x=limit_21, color = 'red')
plt.axvline(x=limit_22, color = 'red')
plt.xticks([np.min(x2), limit_21,limit_22, np.max(x2)])
plt.ylabel('probability')
plt.xlabel('readings')
plt.title('Sensor 2 Distribution')
plt.show()
```

```python
sns.lineplot(x = x3, y= y3)
plt.axvline(x=mean_x3, color = 'black')
plt.axvline(x=limit_31, color = 'red')
plt.axvline(x=limit_32, color = 'red')
plt.xticks([np.min(x3), limit_31,limit_32, np.max(x3)])
plt.ylabel('probability')
plt.xlabel('readings')
plt.title('Sensor 3 Distribution')
plt.show()
```
